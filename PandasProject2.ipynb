#### a) Which state/territory had the highest numerical increase in confirmed cases over the summer months (June 1st through August 31th)

#####  Graphics:  Generate Bar chart for  Age more than 60 count, total population, unemployment count and COVID numerical value for summer months

```python
import pandas as pd
cvd = pd.read_csv("time_series_covid19_confirmed_US.csv")
cvd
```

##### Group the rows by State and get the sum of cases for all states

```python
cvdStates = cvd.groupby("Province_State").sum()
cvdStates
```

##### Filter the columns - Summer months


###### Taking May 1st for getting the June difference value

```python
cvdDF = pd.DataFrame(cvdStates)
cvdSummerMonthsDF = cvdDF[["5/1/20","6/1/20","7/1/20","8/1/20","9/1/20"]]
cvdSummerMonthsDF
```

##### Find the difference between summer for all states

```python
cvdThisPeriodDF = cvdSummerMonthsDF.diff(axis=1);
cvdThisPeriodDF
```

##### Drop NaN values

```python
cvdThisPeriodDF = cvdThisPeriodDF.dropna(axis='columns')
cvdThisPeriodDF
```

##### Find the maximum numerical value for each month

```python
cvdMaxPeriod = cvdThisPeriodDF.max()
cvdMaxPeriod = pd.DataFrame(cvdMaxPeriod, columns = ["Total"])
cvdMaxPeriod
```

##### Find the  states for each month's maximum value 

```python
maxValueIndexObj = cvdThisPeriodDF.idxmax()
maxValueIndexObj
```

##### Combine the Total numerical value and states

```python
cvdMaxPeriod = cvdMaxPeriod.assign(State = maxValueIndexObj)
cvdMaxPeriod
```

##### Get the State list 

```python
states = cvdMaxPeriod['State'].to_list()
print(states)
```

##### Get months list 

```python
months = cvdMaxPeriod.index.values.tolist()
months
```

##### Get numerical value List

```python
numericalValues = cvdMaxPeriod['Total'].to_list()
numericalValues
```

##### Append Month and state for labeling the bar chart

```python
monthAndState = list()
for i in range(0, len(months)):
    monthAndState.append(months[i] + " " + states[i])
```

#### Section 2.
    a) What proportion of cases nationally had occurred in NY State at the start of each month, from March 2020 to August 2020?
    
#####  Graphics:  Generate the circlular graph to show the proportion of NY to nationwide    

```python
cvdStatesDF = pd.DataFrame(cvdStates)
cvdMarToAugDF = cvdStatesDF[["3/1/20","4/1/20","5/1/20","6/1/20","7/1/20","8/1/20"]]
cvdMarToAugDF
```

##### Find the total for all states on these months 

```python
cvdTotalState = cvdMarToAugDF.sum()
cvdTotalState
```

##### Find the total for NY on these months 

```python
cvdNYTotal = cvdMarToAugDF.loc["New York"]
cvdNYTotal
```

##### Find the propertion for NY state

```python
cvdNYProportion = cvdNYTotal.div(cvdTotalState)
print(cvdNYProportion)
```

```python
import seaborn as sns
proprtionResult = pd.DataFrame(cvdTotalState, columns = ["Nationwide Total"])
proprtionResult = proprtionResult.assign(NewYorkTotal = cvdNYTotal)

proprtionResult.assign(Proportion = cvdNYProportion)

print("******************************************************************************************************************");
print("Proportion of cases nationally had occurred in NY State at the start of each month, from March 2020 to August 2020");
print("******************************************************************************************************************");

cm = sns.light_palette("blue", as_cmap=True)
proprtionResult.style.background_gradient(cmap=cm).set_precision(2).highlight_min(axis=0, color='grey').highlight_max(axis=0, color='Orange')
```

##### Get the months and proprtion list to create circular graph

```python
cvdNYProportionDF = pd.DataFrame(cvdNYProportion, columns=["proportion"])
cvdNYProportionDF
```

```python
nyMonths = cvdNYProportionDF.index.values.tolist()
nyMonths
```

```python
proportions = cvdNYProportionDF['proportion'].to_list()
print(proportions)
```

##### Format the proportions to float to draw circle

```python
for i in range(len(proportions)):
    proportions[i]="{0:.2f}".format(proportions[i])
    proportions[i] = float(proportions[i])
    print(proportions[i])
```

##### Approach 1 : Generate the circles to show the proportion of NY to nationwide

```python
import matplotlib.path as mpath
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

for i in range(len(proportions)):
    nyFig = plt.figure()
    ax = nyFig.add_subplot(1, 1, 1)

    r1 = 0.50
    r2 = r1 * (proportions[i])
    NYTitle = "Nationwide NY proprtion for " +  str(nyMonths[i])
    plt.title(NYTitle)
    plt.text(0.2, 0.12,'Nation')
    
    if r2 != 0:        
        plt.text(0.5, 0.5,'NY')
    pp1 = plt.Circle((0.5, 0.5), r1)   
    pp2 = plt.Circle((0.5, 0.5), r2, color='yellow')
   
    ax.add_patch(pp1)
    ax.add_patch(pp2)
    figures.append(nyFig)
```
